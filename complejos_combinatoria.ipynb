```python
def permutacion(palabra):
    if len(palabra) == 1:
        return [palabra]
    perms = permutacion(palabra[1:])
    char = palabra[0]
    result = []
    for perm in perms:
        for i in range(len(perm) +1):
            result.append(perm[:i] + char + perm[i:])
            
    return result
```

```python
x = '123456'
grupo = permutacion(x)
grupo
```
